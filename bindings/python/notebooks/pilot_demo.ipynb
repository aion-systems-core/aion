# AION Python pilot

Requires `AION_LIB_PATH` and `pip install` of the `aion-python` wheel from `bindings/python/`.


```python
import json
import os
from pathlib import Path

from aion import Pilot

root = Path(__file__).resolve().parents[1]
examples = root / "examples"
cap = examples / "demo_capsule_1.json"
ev = examples / "demo_evidence.json"

p = Pilot()
r = p.replay(str(cap))
print(json.dumps({"stdout": r.stdout, "exit_code": r.exit_code}, sort_keys=True))
e = p.evidence(str(ev))
print(json.dumps({"evidence_valid": e.valid}, sort_keys=True))
```
