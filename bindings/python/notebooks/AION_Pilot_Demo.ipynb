# AION-OS Pilot Demo

**Deterministic AI Execution OS** — same capsule, same replay, same hash.

Prerequisites: `maturin develop --release` from `bindings/python/` (or install wheel).

```python
import tempfile
import aion

cap = aion.run("notebook pilot prompt", model="demo", seed=99)
print(cap.model, cap.seed)
path = tempfile.NamedTemporaryFile(suffix=".aionai", delete=False).name
aion.capsule_save(cap, path)
rep = aion.replay(path)
print("replay_symmetry_ok", rep.replay_symmetry_ok)
print("hash", rep.deterministic_hash_hex[:24], "...")
```
